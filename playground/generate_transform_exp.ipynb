```python
# for each vehicle, generate a R and T and a TF matrix per frame

# frame_wise_vehicle_tfs = {} # {frame_id:{veh_id:tf_mat}}
frame_wise_vehicle_tf = defaultdict(defaultdict)
```

```python
# For each frame, for each vehicle, get the transform and generate the required feature map
# 2D R,T matrix 
def get_mat(x,y,t):
    rot = np.array([[cos(-t),-sin(-t),0],[sin(-t),cos(-t),0],[0,0,1]])
    trans = np.array([[1,0,-x],[0,1,-y],[0,0,1]])
    return rot@trans # first translate then rotate
```

```python
for agent_id in agent_wise_idxs:
    agent_df = df.iloc[agent_wise_idxs[agent_id]]
    for idx,row in agent_df.iterrows():
        frame_id = row.frame_id
        frame_wise_vehicle_tf[frame_id][agent_id]={
            "mat": get_mat(row.pos_x,row.pos_y,row.yaw),
            "x":row.pos_x,"y":row.pos_y,"z":row.pos_z,"roll":row.roll,"pitch":row.pitch,"yaw":row.yaw
        }

    
```

```python
# for frame_id in frame_wise_vehicle_tf:
#     for vehicle_id in frame_wise_vehicle_tf[frame_id]:
        
# For each vehicle, store a frame_wise map of all other vehicles(Transformed positions)
# as round_0_vehicle_0.csv
def transform_frame_df(frame_df,frame_id,center_agent_id,save_file=True,round_name="round_0"):
    mat = frame_wise_vehicle_tf[frame_id][agent_id]["mat"]
    frame_df_local = frame_df.copy()
    xy_hmg = frame_df_local[['pos_x','pos_y']].copy()
    xy_hmg['z'] = 1 # creating homogenous coordinate
    frame_df_local[['pos_x','pos_y']]=(mat@xy_hmg.T).T.values[:,:2]
    if save_file:
        frame_df_local.to_csv(f"agent_maps/{round_name}_vehicle_{center_agent_id}_frame_{frame_id}.csv")
    
#     (q@w@df_test.T).T.values[:,:2]
    
    
    

for agent_id in tqdm(agent_wise_idxs):
    agent_df = df.iloc[agent_wise_idxs[agent_id]]
    frame_ids = list(agent_df.frame_id)
    for frame_id in frame_ids:
        frame_df = df[df.frame_id==frame_id] # get full df for the given frame
        transform_frame_df(frame_df,frame_id,agent_id)
```
